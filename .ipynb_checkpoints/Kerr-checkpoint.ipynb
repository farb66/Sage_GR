```python
#set Up
%display latex
import time
comput_time0 = time.perf_counter()
#Parallelism().set(nproc=8)#number of parallel threads
#exper9iments with parallelism
```

```python
M = Manifold(4, 'M', latex_name=r'\mathcal{M}', structure='Lorentzian')
BL.<t,r,th,ph> = M.chart(r"t r th:(0,pi):\theta ph:\phi")#better describes bodies rotating
```

```python
var('m, a', domain='real')
m_val = 2
a_val = 1
'''
> m = mass of the body
> a = angular velocity of the body
'''
```

```python
g = M.metric()

# Maps:
#   From local to global reference frame
#   Cartesian to Boyer-Lindquist Spherical coordinates 
```

```python
#rotation functions
rho2 = r^2 + (a*cos(th))^2 
Delta = r^2 -2*m*r + a^2
'''
> rho2: squate of the distance a points rotates horizontally
> a*cos(th): the distance rotated on the x axis'
'''
```

```python
"----METRIC DEFITION----"
```

```python
g[0,0] = -(1-2*m*r/rho2)
print("----METRIC AT [t, t]---- \n > Lorentian Time Transform adpated \n > C is implicit \n > Rho2 conncted with t2 is local angular horizontal velocity")
```

```python
g[0,3] = -2*a*m*r*sin(th)^2/rho2
print("----METRIC AT [t, phi]---- \n > Change rate in proper time Horizontal \n > Proportional to: \n - vertical torsion \n - mass being rotated")
```

```python
g[1,1]= rho2/Delta
print("----METRIC AT [r, r]---- \n > Spacial Contraction \n > radial distance change")
```

```python
g[2,2] = rho2
print("----METRIC AT [th, th]---- \n > Horizontal Angular shift")
```

```python
g[3,3] = (r^2+a^2+2*m*r*(a*sin(th))^2/rho2)*sin(th)^2
print("----METRIC AT [phi, phi]---- \n > Vertical Angular shift")
```

```python
print("------------Kretschmann Scalar------------")
```

```python
R = g.riemann()
dR = R.down(g)
uR = R.up(g)
Kr_scalar = uR['^{abcd}']*dR['_{abcd}']
Kr_scalar.display()
```

```python
Kr = Kr_scalar.coord_function()
K1 = Kr.expr().subs(m=0.3, a=0.9)
plot3d(K1, (r, 1, 2), (th, 0, pi), axes_labels=['r', 'theta', 'Kr'])
```

```python
print("------------Single Geodesic Image------------")
```

```python
p = M((0, 15, pi/2, 0))#manifold subsection limits
Tp = M.tangent_space(p)
v = Tp((2, 0, 0.005, 0.05))#subsection tangentspace
v = v / sqrt(-g.at(p)(v, v))
```

```python
E.<x, y, z> = EuclideanSpace()
phi = M.diff_map(E, [r*sin(th)*cos(ph), r*sin(th)*sin(ph), r*cos(th)])
phi.display()
```

```python
tau = var('tau')#geodesic 
```

```python
curve = M.integrated_geodesic(g, (tau, 0, 5000), v)
sol = curve.solve(step = 1, method="ode_int", parameters_values={m: m_val, a: a_val})#body
```

```python
interp = curve.interpolate()

P = curve.plot_integrated(mapping=phi, color="red", thickness=2, plot_points=5000)
P += sage.plot.plot3d.shapes.Sphere(4, color='grey')
P
```
